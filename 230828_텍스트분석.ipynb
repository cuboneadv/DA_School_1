```python
import pandas as pd

# 데이터를 pandas DataFrame으로 불러옴
review_data = pd.read_excel("/content/review_20230725_091330_da.xlsx")
```

```python
review_data
```

```python
import sqlite3

# 1. SQLite3를 사용하여 데이터베이스 파일 생성
conn = sqlite3.connect("reviews.db")

# 2. 해당 데이터베이스에 테이블을 생성하고 데이터를 입력
review_data.to_sql('product_reviews', conn, if_exists='replace', index=False)

# 3. 데이터를 조회하여 확인
query = "SELECT * FROM product_reviews LIMIT 5"
result = pd.read_sql(query, conn)
result
```

```python
query = """
  SELECT 상품명, 리뷰상세내용
  FROM product_reviews
  LIMIT 10
"""
result = pd.read_sql(query, conn)
result
```

```python
query = """
    SELECT 상품명, COUNT(리뷰상세내용) AS 리뷰개수
    FROM product_reviews
    GROUP BY 상품명
    ORDER BY 리뷰개수 DESC
"""
result = pd.read_sql(query, conn)
result
```

```python
query = """
    SELECT 상품명, COUNT(리뷰상세내용) AS 리뷰개수, AVG(구매자평점)
    FROM product_reviews
    WHERE 상품명 LIKE '%쥬시%' AND 리뷰상세내용 LIKE '%맛있다%'
"""
result = pd.read_sql(query, conn)
result
```

```python
query = """
    SELECT
        CASE WHEN 리뷰상세내용 LIKE '%쥬시%' THEN '쥬시'
             WHEN 리뷰상세내용 LIKE '%아메리카노%' THEN '아메리카노'
             WHEN 리뷰상세내용 LIKE '%에스프레소%' THEN '에스프레소'
             WHEN 리뷰상세내용 LIKE '%드립백%' THEN '드립백'
             WHEN 리뷰상세내용 LIKE '%캡슐%' THEN '캡슐'
             ELSE '기타'
             END AS 커피_종류,
        AVG(구매자평점) AS 평균_평점
    FROM product_reviews
    GROUP BY 커피_종류
    ORDER BY AVG(구매자평점) DESC
"""
result = pd.read_sql(query, conn)
result

# CASE WHEN 구문 사용 시 주의할 점: 한 구문에 두 가지 이상의 단어가 포함될 수도 있음
# 예) '쥬시 아메리카노 너무 맛있어요'의 경우, 쥬시, 아메리카노가 한 구문에 모두 포함되어 있음
```
