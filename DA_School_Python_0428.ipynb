* 대부분의 코드는 for로 대체할 수 있습니다.
* 효율도 매우 좋지 않습니다. 다만 메모라이징 기법을 사용하면 좋긴 합니다.
* 다이나믹 프로그래밍 등 필수적으로 사용되는 곳이 있기도 합니다.

```python
# sys.setrecursionlimit -> 제한을 늘릴 수 있음
def 숫자출력():
    print(1)
    return 숫자출력()
숫자출력()
```

```python
def factorial(n):
    if n <= 1:
        return n
    return n * factorial(n-1)

factorial(5)

#                   [return 값]
# factorial(5) == 5 * factorial(4) == 5 * 24
# factorial(4) == 4 * factorial(3) == 4 * 6
# factorial(3) == 3 * factorial(2) == 3 * 2
# factorial(2) == 2 * factorial(1) == 2 * 1
# factorial(1) == 1
```

```python
# for문을 이용할 수 있는데 굳이 factorial 함수를 이용할 필요 없음
result = 1
for i in range(1, 6):
    result *= i

result
```

```python
def string_reverse(s):
    if len(s) == 1:
        return s
    return string_reverse(s[1:]) + s[0]

string_reverse('hello')

# string_reverse('hello') == string_reverse('ello') + 'h' == 'olle' + 'h' == 'olleh'
# string_reverse('ello') == string_reverse('llo') + 'e' == 'oll' + 'e' == 'olle'
# string_reverse('llo') == string_reverse('lo') + 'l' == 'ol' + 'l' == 'oll'
# string_reverse('lo') == string_reverse('o') + 'l' == 'o' + 'l' == 'ol'
# string_reverse('o') == 'o'
```

```python
# 재귀가 얼마나 비효율적인지 보여주는 사례
def fib(num):
    if num == 0:
        return 0
    if num == 1 or num == 2:
        return 1
    else:
        return fib(num-1) + fib(num-2)

fib(35)
```

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def fib(num):
    if num == 0:
        return 0
    if num == 1 or num == 2:
        return 1
    else:
        return fib(num-1) + fib(num-2)

fib(40)
```

```python
def plus(num1, num2):
    try:
        result = int(num1) + int(num2)
        return result
    except:
        return '연산할 수 없습니다'

print(f'plus: {plus(10, 5)}')
print(f'plus: {plus("10", 5)}')
print(f'plus: {plus("10", "5")}')
print(f'plus: {plus(10, "5")}')
print(f'plus: {plus("abc", 5)}')
print(f'plus: {plus(10, "abc")}')
```

```python
plus = lambda x, y: x + y
minus = lambda x, y: x - y
multiply = lambda x, y: x * y
divide = lambda x, y: x / y

print(f'plus : {plus(10, 5)}')
print(f'minus : {minus(10, 5)}')
print(f'multiply : {multiply(10, 5)}')
print(f'divide : {divide(10, 5)}')
```

```python
# 3. x의 n제곱을 구하는 함수를 만들어주세요.
#    재귀함수를 이용하여 만들어야하며, x는 0이 아닙니다.
def 제곱(x, n):
    if n == 0:
        return 1
    return x * 제곱(x, n - 1)

제곱(3, 4)
```

```python
# factory 함수
# 클로저
def factory(x):
    def 승수(n):
        return x ** n
    return 승수

제곱3 = factory(3)
# 실제 제곱3은 아래와 같습니다.
# def 승수(n):
#     return 3 ** n

제곱4 = factory(4)

제곱3(2)
제곱3(3)
제곱4(2)
제곱4(4)
```
