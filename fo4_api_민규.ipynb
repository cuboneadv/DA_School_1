```python
import time

matchtype = match_type_df['matchtype'].to_list()[5:]
offset = 0  # 569993700
limit = 100
orderby = 'desc'

matches = []

for m_type in matchtype:
    print(m_type, type(m_type))
    url = f'https://api.nexon.co.kr/fifaonline4/v1.0/matches?matchtype={m_type}&offset={offset}&limit={limit}&orderby={orderby}'
    # print(get_raw_data(url))
    match_id = get_raw_data(url)
    matches.extend(match_id)
    # time.sleep(5)
```

```python
import json

puuids = []

for matchid in matches:
    print(matchid)
    try:
        url = f'https://api.nexon.co.kr/fifaonline4/v1.0/matches/{matchid}'
        puuid = pd.json_normalize(get_raw_data(url)['matchInfo'])['accessId'].to_list()
        # print(puuid)
        # print(json.dumps(match_detail_json, indent = 4))
        puuids.extend(puuid)
    except:
        pass
    # break
```

```python
match_types = []
cnt = 1

for accessid in puuids:
    try:
        url = f'https://api.nexon.co.kr/fifaonline4/v1.0/users/{accessid}/maxdivision'
        matchtype = pd.json_normalize(get_raw_data(url))['matchType'].to_list()
        # print(matchtype)
        match_types.extend(matchtype)
        cnt += 1
        # break
        if cnt % 10 == 0:
            time.sleep(10)
    except:
        pass
    # break
```

```python
def connect_RDS(host, port, username, password, database):
    try :
        conn = pymysql.connect(host=host,
                               user=username,
                               passwd=password,
                               db=database,
                               port=port,
                               use_unicode=True,
                               charset='utf8')
        cursor = conn.cursor()

    except :
        logging.error("RDS에 연결되지 않았습니다.")
        sys.exit(1)

    return conn, cursor
```
