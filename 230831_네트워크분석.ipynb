```python
import networkx as nx
import matplotlib.pyplot as plt

# 그래프 생성
G = nx.Graph()

# 노드 추가
G.add_node("A")
G.add_node("B")

# 엣지 추가
G.add_edge("A", "B")

# 그래프 시각화
nx.draw(G, with_labels=True)
plt.show()
```

```python
# import networkx as nx

G = nx.Graph()  # 그래프 생성

G.add_node(1)  # 노드 추가
G.add_nodes_from([2, 3])  # 여러 노드를 한 번에 추가

G.add_edge(1, 2)  # 엣지 추가
G.add_edges_from([(2, 3), (3, 1)])  # 여러 엣지를 한 번에 추가
```

```python
# 그래프 시각화
plt.figure(figsize=(6, 6))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1000, edge_color='gray', font_size=15)
plt.title("Simple Graph")
plt.show()
```

```python
# 방향성 그래프 (DiGraph) 생성 및 노드, 엣지 추가
DG = nx.DiGraph()
DG.add_node(1)
DG.add_nodes_from([2, 3])
DG.add_edge(1, 2)
DG.add_edges_from([(2, 3), (3, 1)])

# 그래프 시각화
plt.figure(figsize=(6, 6))
nx.draw(DG, with_labels=True, node_color='lightgreen', node_size=1000, edge_color='gray', font_size=15, arrowsize=20)
plt.title("Directed Graph (DiGraph)")
plt.show()
```

```python
# 그래프 생성 및 노드 추가
G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2, 3])
G.add_nodes_from([
    (4, {"color": "red"}),
    (5, {"color": "green"}),
])

# 노드 속성에 따른 색상 지정
node_colors = [G.nodes[node].get("color", "blue") for node in G.nodes()]

# 그래프 시각화
plt.figure(figsize=(7, 7))
nx.draw(G, with_labels=True, node_color=node_colors, node_size=1000, edge_color='gray', font_size=15)
plt.title("Graph with Node Attributes")
plt.show()
```

```python
# 경로 그래프 생성 및 그래프 G에 노드 추가
H = nx.path_graph(10)
G.add_nodes_from(H)

# 노드 속성에 따른 색상 지정 (기존 코드 사용)
node_colors = [G.nodes[node].get("color", "blue") for node in G.nodes()]

# 그래프 시각화
plt.figure(figsize=(8, 8))
nx.draw(G, with_labels=True, node_color=node_colors, node_size=1000, edge_color='gray', font_size=15)
plt.title("Extended Graph with Path Graph Nodes")
plt.show()
```

```python
# 그래프 H를 그래프 G의 노드로 추가
G.add_node(H)

# 그래프 시각화
plt.figure(figsize=(10, 10))
nx.draw(G, with_labels=True, node_color=node_colors, node_size=1000, edge_color='gray', font_size=15)
plt.title("Graph with Path Graph as a Node")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 데이터 불러오기
G = nx.karate_club_graph()

# 기본 정보 확인
print(G)

# 네트워크 시각화
nx.draw(G, with_labels=True)
plt.show()
```

```python
# 연결 중심성 계산 및 시각화
degree_centrality = nx.degree_centrality(G)
nx.draw(G, with_labels=True, node_size=[v * 5000 for v in degree_centrality.values()])
plt.title("Degree Centrality")
plt.show()
```

```python
# Louvain 방법을 이용한 커뮤니티 탐지는 community 모듈을 활용
from community import community_louvain

# 커뮤니티 탐지
partition = community_louvain.best_partition(G)

# 결과 시각화
nx.draw(G, with_labels=True, node_color=list(partition.values()))
plt.title("Detected Communities")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.florentine_families_graph()
plt.figure(figsize=(10, 8))
nx.draw(G, with_labels=True, node_size=3000, node_color="lightblue", font_size=15)
plt.title("Florentine Families Network", fontsize=20)
plt.show()
```

```python
# 데이터 불러오기 및 시각화
G_florentine = nx.florentine_families_graph()

plt.figure(figsize=(10, 8))
nx.draw(G_florentine, with_labels=True, node_size=3000, node_color="lightblue", font_size=15)
plt.title("Florentine Families Network", fontsize=20)
plt.show()

# 데이터 탐색: 가장 많은 결혼 관계를 가진 가문 확인
degree_sequence_florentine = sorted(G_florentine.degree(), key=lambda x: x[1], reverse=True)

# 네트워크 밀도 계산
density_florentine = nx.density(G_florentine)

degree_sequence_florentine[:5], density_florentine
```
